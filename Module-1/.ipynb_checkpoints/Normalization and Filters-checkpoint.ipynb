```python
%matplotlib inline
```

```python
# load the image from source
image_file = "test.jpg"
image = mp_image.imread(image_file)
plt.imshow(image)
```

# Image Normalization

Let's look at the distribution of pixel values in image. Ideally, the image should have relatively even distribution indicating good contrast and making it easier to extract analytical information. 

Let's plot a histogram.

```python
# We need to use ravel to flatten the 3 dimensions
plt.hist(image.ravel(), bins=255, cumulative=True)
plt.show()
```

Contrast stretching, in which we rescale the pixel values b/w a very low and very high pecentile. <bold>Histogram equalization</bold> is a more complex operation that creates a more uniform distribution. The code in the cell below uses the <bold>exposure.rescale_intensity</bold> and <bold>exposure.equalize_hist</bold> methods from the skimage package to perform contrast stretching and histogram equalization.

```python
!pip install --upgrade scikit-image
```

```python
from skimage import exposure
import numpy as np
```

```python
p2 = np.percentile(image, 2)
p98 = np.percentile(image, 98)

image_ct = exposure.rescale_intensity(image, in_range=(p2, p98))

# Histogram equalization
image_eq = exposure.equalize_hist(image)

fig = plt.figure(figsize=(20,12))
a = fig.add_subplot(3,3,1)
imgplot = plt.imshow(image)
a.set_title('Original')
```

```python
a = fig.add_subplot(3,3,2)
imgplot = plt.imshow(image_ct)
a.set_title('Contrast stretched')
```

```python
a = fig.add_subplot(3,3,3)
imgplot = plt.imshow(image_eq)
a.set_title('Histogram equalized')
```

```python
a = fig.add_subplot(3,3,4)
imgplot = plt.hist(image.ravel())
```

```python
a=fig.add_subplot(3,3,5)
imgplot=plt.hist(image_ct.ravel())
```

```python
a=fig.add_subplot(3,3,6)
implot=plt.hist(image_eq.ravel())
```

```python
# Subplots for CDFs

a=fig.add_subplot(3,3,7)
imgplot = plt.hist(image.ravel(), bins=255, cumulative=True)
```

```python

a=fig.add_subplot(3,3,8)
imgplot = plt.hist(image_ct.ravel(), bins=255, cumulative=True)
```

```python

a=fig.add_subplot(3,3,9)
imgplot = plt.hist(image_eq.ravel(), bins=255, cumulative=True)
```

```python
plt.show()
```

As you can see, contrast stretching results in a histogram of more or less the same shape, but spread out over the full range of pixel values. The CDF is slightly smoother than for the original image. Histogram equalization has produced a much more even distribution of pixel values, and a more or less perfectly diagonal CDF; but the image itself is noticable less natural looking.

These kind of contrast enhancements can be useful when images have low contrast. They're particularly useful in medical imaging scenarios, for example to enhance X-rays. They can also be used to counteract contrast variance in a set of images that will be used to train a machine learning model.

# Filters

Fliters are implemented by defining numeric grids, called kernels that are convolved across an image to change the middle of the grid by calculating a weighted sum of the surrounding pixels using the the values in the kernel. They are used to apply visual enhancements to image, such as sharpening, blurring and so on. This remove "noise" from an image, such as is common in photograph taken in low light conditions.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
```

```python
%matplotlib inline
```

```python
image_file="test.jpg"
image = Image.open(image_file)
```

```python
edges_image = image.filter(ImageFilter.BoxBlur(10))
```

```python
fig = plt.figure(figsize=(16,12))
a = fig.add_subplot(1,2,1)
image_plot_1= plt.imshow(image)
a.set_title("Original")
```

```python
# Plot filtered image
a = fig.add_subplot(1,2,1)
image_plot_2=plt.imshow(edges_image)
a.set_title("Edges")
plt.show()
```
